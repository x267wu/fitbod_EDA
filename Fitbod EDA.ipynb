# Collaborative filtering model
For the machine learning portion of this problem, I will attempt to build and evaluate a recommender system. 
The use-case for this system would be to recommend exercises to workouts. This could guide new gym enthusiasts in the process of building their first workout, or add a few extra loads to an existing workout to really pump those muscles.

## Model description
The data will be multi-hot encoded to count unique occurences of exercises within a workout and passed to a KNearestNeighbor Classifier. We will be using cosine distance, such that we can get the cosine similarities and create a weighted average decision. We then return all exercises that are rated above a certain threshold as recommendations. If no exercise are above a threshold, no recommendation is made (this could be changed based on use cases).
We are basically doing user-based collaborative filtering, where a workout is considered the user, and exercises are the items.

```python
import numpy as np
from sklearn.metrics import auc
from sklearn.neighbors import NearestNeighbors
from collections import Counter
from scipy import spatial
from math import sqrt
import matplotlib.pyplot as plt

def cosine_similarity(a,b):
    return 1 - spatial.distance.cosine(a, b)
    
# helper function to transform data into multi-hot encoding
def multi_hot(df, length = 101, unique_counts = True, append_user_id = True):
    if unique_counts == True:
        counts_for_exercise = Counter(list(df["exerciseIndex"].unique())) #we only look at unique counts
    else:
        counts_for_exercise = Counter(list(df["exerciseIndex"]))
    multi_hot_list = [0] * length
    for exercise in counts_for_exercise.keys():
        multi_hot_list[exercise] = float(counts_for_exercise[exercise])
    if append_user_id:
        multi_hot_list.append(df["userIndex"].unique()[0])
    return np.asarray(multi_hot_list)

def prediction_results(threshold, training_set, test_set, K, folds):
    false_positives = 0
    false_negatives = 0
    true_positives = 0
    true_negatives = 0
    if folds > 1:
        X = np.concatenate((training_set, test_set))
        fold_size = len(test_set)
    for f in range(folds):
        if folds > 1:
            test_set = X[fold_size * f:fold_size * (f+1)]
            training_set = np.concatenate((X[:fold_size * f], X[(f+1) * fold_size:]))
        for i in range(len(test_set[0])):
            idx_OUT_columns = [i]
            idx_IN_columns = [i for i in range(np.shape(training_set)[1]) if i not in idx_OUT_columns]
            neigh = NearestNeighbors(n_neighbors=K, metric='cosine', algorithm='brute')

            # we do not train with feature i
            neigh.fit(np.stack(training_set[:,idx_IN_columns]))
            iteration_FP, iteration_FN, iteration_TP, iteration_TN = count_hits_and_misses(i, training_set, test_set[:,idx_IN_columns], neigh, threshold, test_set[:,i])
            #neigh.fit(np.stack(training_set))
            #iteration_FP, iteration_FN, iteration_TP, iteration_TN = count_hits_and_misses(i, training_set, test_set, neigh, threshold, test_set[:,i])
            false_positives += iteration_FP
            false_negatives += iteration_FN
            true_positives += iteration_TP
            true_negatives += iteration_TN
    return true_positives, false_positives, true_negatives, false_negatives


# This counts the prediction resuts
def count_hits_and_misses(exercise_index, training_set, test_set, knn_model, threshold, true_labels):
    false_positives = 0
    false_negatives = 0
    true_positives = 0
    true_negatives = 0
    # find k nearest workouts
    cosine_distances, indices_of_most_similar_workouts = knn_model.kneighbors(test_set, return_distance=True)
    for i, workout in enumerate(test_set):
        workouts_neighbours_subset = training_set[indices_of_most_similar_workouts[i]]
        # similarities = 1 - cosine distance
        averaged_ratings = np.ma.average(workouts_neighbours_subset, weights=cosine_distances[i] * (-1) + 1, axis= 0)
        
        # calculate thresholded label prediction
        predicted_label = averaged_ratings[exercise_index] > threshold
        
        if predicted_label == (true_labels[i] > 0): #if prediction is correct
            if predicted_label: #if recommended
                true_positives +=1
            else:
                true_negatives +=1
        else:
            if predicted_label:
                false_positives +=1
            else:
                false_negatives +=1
    return false_positives, false_negatives, true_positives, true_negatives

# Calculate Matthew's Correlation's Coefficient (better for imbalanced datasets than F1)
def calculate_MCC(TP, FP, TN, FN):
    return (TP * TN - FP * FN) / sqrt((TP+FP)* (TP+FN)*(TN+FP)*(TN+FN))

# Calculate metrics of fit for a set of prediction results
def calculate_metrics_of_fit(TP, FP, TN, FN):
    print("Confusion Matrix:")
    print(np.array([[TP,FP],[FN,TN]]))
    P = TP + FN
    N = TN + FP
    PP = TP + FP
    PN = TN + FN
    print("Total positives (exercise performed): " + str(P))
    print("Total negatives (exercise not performed): " + str(N))
    print("False positives (exercise recommended, but not performed): " + str(FP))
    print("False negatives (exercise not recommended, but actually performed): " + str(FN))
    print("True positives (exercise recommended and performed): " + str(TP))
    print("True negatives (exercise not recommended and not performed): " + str(TN))
    accuracy = (TP + TN) / (P + N)
    precision = TP/PP
    recall = TP / P
    fallout = FP/N
    MCC = calculate_MCC(TP, FP, TN, FN)
    F1 = 2 * precision * recall / (precision + recall)
    print("accuracy: " + str(accuracy))
    print("precision: " + str(precision))
    print("recall: " + str(recall))
    print("MCC: " + str(MCC))
    print("F1: " + str(F1))
    print("\n")

    return MCC, recall, fallout, precision, accuracy, F1

def train_model(training_set, test_set, validation_set, threshold_range, minimum_distinct_exercises = 1, K = 150, folds = 1, eval_metric = "MCC"):
    test_set = test_set[np.sum(test_set,axis=1) >= minimum_distinct_exercises]
    validation_set = validation_set[np.sum(validation_set,axis=1) >= minimum_distinct_exercises]
    
    recalls = []
    fallouts = []
    MCCs = []
    F1s = []
    # test prediction results for each threshold.
    threshold_step_size = threshold_range[1] - threshold_range[0]
    for threshold in threshold_range:
        print("Testing threshold", threshold)
        true_positives, false_positives, true_negatives, false_negatives = prediction_results(threshold, training_set, test_set, K, folds)
        try:
            MCC, recall, fallout, _, _, F1 = calculate_metrics_of_fit(true_positives, false_positives, true_negatives, false_negatives) 
            recalls.append(recall)
            fallouts.append(fallout)
            MCCs.append(MCC)
            F1s.append(F1)
        except:
            break

    if eval_metric == "MCC":
        optimal_threshold = np.argmax(MCCs)*threshold_step_size
        print("optimal MCC: ", np.max(MCCs) , "at threshold=", optimal_threshold)
    elif eval_metric == "F1":
        optimal_threshold = np.argmax(F1s)*threshold_step_size
        print("optimal F1: ", np.max(F1s) , "at threshold=", optimal_threshold)
    if folds > 1: #use test set for validation training
        true_positives, false_positives, true_negatives, false_negatives = prediction_results(optimal_threshold, np.concatenate((training_set, test_set)), validation_set, K,1)
    else:
        true_positives, false_positives, true_negatives, false_negatives = prediction_results(optimal_threshold, training_set, validation_set, K, 1)  
    print("validating...")
    
    MCC, recall, fallout, precision, accuracy, F1 = calculate_metrics_of_fit(true_positives, false_positives, true_negatives, false_negatives) 
    P = true_positives + false_negatives
    N = true_negatives + false_positives
    print("accuracy of null model (no recommendations): ", N / (P + N))
    #run plots after
    plt.plot(np.arange(len(MCCs)) * threshold_step_size, MCCs)
    plt.title("Matthew's Correlation Coefficient by Threshold")
    plt.xlabel("Decision Threshold")
    plt.ylabel("MCC")
    plt.show()
    
    plt.plot(fallouts,recalls)
    plt.title("ROC Curve")
    plt.xlabel("Fallout")
    plt.ylabel("Recall")
    plt.show()

    plt.plot(np.arange(len(F1s)) * threshold_step_size, F1s)
    plt.title("F1 by Threshold")
    plt.xlabel("Decision Threshold")
    plt.ylabel("F1")
    plt.show()
    
    return {"optimal MCC": np.max(MCCs) ,"optimal decision threshold" :optimal_threshold, "validation MCC" : MCC, "validation recall": recall, "precision" : precision, "accuracy" : accuracy, "accuracy of null model (no recommendations)" : N / (P + N)}

```

```python
exerciseDict = {b: a for a, b in enumerate(df["exerciseName"].unique())}
df["exerciseIndex"] = df["exerciseName"].apply(lambda x: exerciseDict[x])
userDict = {b: a for a, b in enumerate(df["fake_user_id"].unique())}
df["userIndex"] = df["fake_user_id"].apply(lambda x: userDict[x])
workout_exercise_matrix = np.asarray(df.groupby("workoutId").apply(lambda x: multi_hot(x)).reset_index(name='multi_hot')['multi_hot'])
workout_exercise_matrix = np.stack(workout_exercise_matrix)
display(workout_exercise_matrix)
```

In order to determine the threshold, we will separate our data into a training set, a test set and a validation set. The test set will be used to tune the threshold whereas the validation set will evaluate the overall performance of the model.

How do we determine our test and training however? It seems like how we split the data set actually affects the nature of our model...

On one hand, if we split the data by randomly sampling amongst the workouts, we might end up with workouts from the same user in both the test set and the training set. If we split the data like this, it would be assumed that the recommender also draws information from a user's past workouts, and recommend exercises based on the user's history. 

On the other hand, if we split the data by randomly sampling amongst the users, we assume the model does not make recommendations based on workouts from the same user.

For this problem, we will try both and compare the results. Moreover, for both methods, we will test what happens when we up the minimum exercise requirement to 2 and 3 instead of 1. This is because collaborative filtering systems are bad at recommending items to new users. If we mask the index for which we have an exercise, we are essentially just asking the model to make a prediction based off no data (find similarities to an array of 0s). Moreover, the data set is biased towards workouts with only 1 exercise (see graph at the begining).

From our results, we can see that the models that are testing by random users perform extremely poorly (validation MCC < 0.2) The correlation is negligeable. This is somewhat logical since, half the time, we are predicting on vectors of `0`. Even when we are not, the the vectors are so sparse that the model can't predict anything interesting. We can see that removing workouts that have only 1 exercise from the test and validation set slightly improves the model, but we still fail to achieve any statistically significant result. The model testing workouts with minimum 3 exercises also fail to reach any statistical significance. 

We can also see that, allowing exercises to be predicted based off of workouts from the same user improves the model as well. This makes sense, since we expect workouts from the same user to have more predictive power, as they are correlated by the user. This does decrease the value of our prediction since the user has most often time already seen the exercise. Indeed, although testing workouts with minimum 1 unique exercise still gives insignificant results, testing workouts with minimum 2 unique exercises and testing workouts with minimum 3 unique exercises provide weak and moderate predictive power respectively (validation MCC = 0.2413 and 
0.3049)

Since a model with weak predictive power is sad, I will explore doing collaborative filtering based on users instead of workouts. Moreover, I will be using 10 fold cross validation for this model (the other model was taking too long to test thresholds -- I could optimize the model but I think its out of scope for this exercise. We should always cross validate.) Another thing introduced in this model is that we use the implicit signals of frequency of exercise as a rating (which we normalize).

```python
from sklearn.preprocessing import normalize
user_exercise_matrix = np.asarray(df.groupby("fake_user_id").apply(lambda x: multi_hot(x, unique_counts = False, append_user_id = False)).reset_index(name='multi_hot')['multi_hot'])
user_exercise_matrix = np.stack(user_exercise_matrix)
normalized_user_exercise_matrix = normalize(user_exercise_matrix, axis = 1, norm='l2')
normalized_user_exercise_matrix

#Random sampling 
np.random.shuffle(normalized_user_exercise_matrix)
training_set = normalized_user_exercise_matrix[:60]
test_set = normalized_user_exercise_matrix[63:68]
validation_set = normalized_user_exercise_matrix[68:]
train_model(training_set, test_set, validation_set, np.arange(0., 0.1, 0.001), K = 8, folds=10, eval_metric="MCC")  
```

In conclusion, we see that collaborative filtering works better on users than on workouts. With an MCC of 0.48705 on the validation set, we can say there is strong positive relationship between the predictors (similar users) and the predicted class.

This is because `user_exercise_matrix` is much less sparse than `workouts_exercise_matrix` (shown below).

This results in a more accurate score of similarities between users, whereas a big portion of workouts have 0 similarities between them. 

However, the user-based filtering model does not answer the questions asked by the workouts-based filtering model. It might make more sense to recommend an exercise within the context of a workout, since a workout has dedicated muscle groups, and we might not want a weak recommendation without taking that context into account. 

We saw that the workouts-based filtering model worked best when workouts from the user were also part of the neighbour space for the KNN Model. This shows the fact that other workouts from a same user have more predictive power than workouts from other users. This shifts the use case of the model, since the recommendations are going to be mostly exercises a user has already seen. Maybe it is part of a "quick exercise builder feature" where we can drag and drop exercises... idk. All in all, even if only the user-based filtering had substantial statistical results, I think all three models are useful, but they each have a different use case.
